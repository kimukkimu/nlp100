## 30. 形態素解析結果の読み込み

```python
import MeCab

input_file = 'neko.txt'
output_file = 'neko.txt.mecab'

# 「吾輩は猫である」を形態素解析し、ファイル出力
def neko_parse():
    with open(input_file) as i, open(output_file, mode='w') as o:
        m = MeCab.Tagger()
        o.write(m.parse(i.read()))

def neko_lines():
    result_list = []
    result_lists = []
    
    with open(output_file) as o:
        for line in o:
            surface_split = line.split('\t')
            if(len(surface_split) < 2):
                continue
            morpheme_split = surface_split[1].split(',')
            
            morpheme = {
                'surface': surface_split[0],
                'base': morpheme_split[6],
                'pos': morpheme_split[0],
                'pos1':morpheme_split[1]
            }
            result_list.append(morpheme)
            
            if morpheme_split[1] == '句点':
                result_lists.append(result_list)
                result_list = []
    return result_lists

morpheme_result = neko_lines()
morpheme_result
```

## 31. 動詞
動詞の表層形をすべて抽出せよ．

```python
for x in morpheme_result:
    for morpheme in x:
        if morpheme['pos'] == '動詞':
            print(morpheme['surface'])
```

## 32. 動詞の原形
動詞の原形をすべて抽出せよ．

```python
for x in morpheme_result:
    for morpheme in x:
        if morpheme['pos'] == '動詞':
            print(morpheme['base'])
```

## 33. サ変名詞
サ変接続の名詞をすべて抽出せよ

```python
for x in morpheme_result:
    for morpheme in x:
        if morpheme['pos1'] == 'サ変接続':
            print(morpheme)
```

## 34. 「AのB」
2つの名詞が「の」で連結されている名詞句を抽出せよ．

```python
answer  = []
for x in morpheme_result:
    A_B_flag = 0
    A_B = ''
    for morpheme in x:
        if morpheme['pos'] == '名詞' and A_B_flag == 0:
            A_B += morpheme['surface']
            A_B_flag = 1
        elif A_B_flag ==1 and morpheme['surface'] == 'の':
            A_B += morpheme['surface']
            A_B_flag = 2
        elif A_B_flag == 2  and morpheme['pos'] == '名詞':
            A_B += morpheme['surface']
            answer.append(A_B)
            A_B_flag = 0
            A_B = ''
        else:
            A_B_flag = 0
            A_B = ''
answer
            
```
